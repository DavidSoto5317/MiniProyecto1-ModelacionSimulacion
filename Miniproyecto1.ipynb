### Ejercicio 1 - El triángulo de Shierpinski

```python
### Se importa la librería random para poder obtener numeros random entre 0 y 1
import random
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
import numpy as np
```

```python
### Se definen las funciones a utilizar
def f1(p):
    return [p[0]/2, p[1]/2]

def f2(p):
    return [p[0]/2 + 0.5, p[1]/2]

def f3(p):
    return [p[0]/2 + 0.25, p[1]/2 + 0.5]
```

```python
### Se hace una funcion que mapee a una de las funciones anteriores segun la probabilidad

### Probabilidades
p1 = 0.1
p2 = 0.2
p3 = 0.7

### Definicion para la funcion que mapea probabilidades
def mapeo(punto, probabilidad):
    if probabilidad > p1 + p2:
        return f3(punto)
    elif probabilidad > p1:
        return f2(punto)
    else:
        return f1(punto)
        
```

### Inciso 1

```python
### Se hace el triangulo de Shierpinski
ITERACIONES = 100000
punto = [0, 0]
arregloPuntosX = [0]
arregloPuntosY = [0]

for i in range(ITERACIONES):
    probabilidad = random.random()
    punto = mapeo(punto, probabilidad)
    arregloPuntosX.append(punto[0])
    arregloPuntosY.append(punto[1])
    
plt.scatter(np.asarray(arregloPuntosX), np.asarray(arregloPuntosY), s=3, c='skyblue')
plt.show()
```

### Inciso 2

```python
### Creamos una funcion para el triangulo parametrizable
def trianguloShierpinski(p1,p2,p3):
    def mapeo(punto, probabilidad):
        if probabilidad > p1 + p2:
            return f3(punto)
        elif probabilidad > p1:
            return f2(punto)
        else:
            return f1(punto)
    return mapeo
```

```python
### Hacemos pruebas para paremtrizar el tringulo por probabilidades
### Probabilidades
p11 = 0.3
p12 = 0.4
p13 = 0.3

funciones = trianguloShierpinski(p11,p12,p13)

### Se hace el triangulo de Shierpinski
ITERACIONES = 100000
punto = [0, 0]
arregloPuntosX = [0]
arregloPuntosY = [0]

for i in range(ITERACIONES):
    probabilidad = random.random()
    punto = funciones(punto, probabilidad)
    arregloPuntosX.append(punto[0])
    arregloPuntosY.append(punto[1])
    
plt.scatter(np.asarray(arregloPuntosX), np.asarray(arregloPuntosY),s=3, c='red')
plt.show()
```

### Ejercicio 2 - El helecho de Barnsley

```python
### Se definen las funciones a utilizar
def f21(p):
    return [p[0]*0.85 + p[1]*0.04 + 0.0, p[0]*(-0.04) + p[1]*0.85 + 1.6]

def f22(p):
    return [p[0]*(-0.15) + p[1]*0.28 + 0.0, p[0]*(0.26) + p[1]*0.24 + 0.44]

def f23(p):
    return [p[0]*0.2 + p[1]*(-0.16) + 0.0, p[0]*(0.23) + p[1]*0.22 + 1.6]

def f24(p):
    return [p[0]*0.0 + p[1]*0.0, p[0]*(0.0) + p[1]*0.16]
```

```python
### Creamos una funcion para el triangulo parametrizable
def helechoBarnsley(p1,p2,p3,p4):
    def mapeo(punto, probabilidad):
        if probabilidad > p1 + p2 + p3:
            return f24(punto)
        elif probabilidad > p1 + p2:
            return f23(punto)
        elif probabilidad > p1:
            return f22(punto)
        else:
            return f21(punto)
    return mapeo
```

```python
### Hacemos pruebas para paremtrizar el tringulo por probabilidades
### Probabilidades
p1 = 0.85
p2 = 0.07
p3 = 0.07
p4 = 0.01

funciones = helechoBarnsley(p1,p2,p3,p4)

### Se hace el triangulo de Shierpinski
ITERACIONES = 100000
punto = [0, 0]
arregloPuntosX = [0]
arregloPuntosY = [0]

for i in range(ITERACIONES):
    probabilidad = random.random()
    punto = funciones(punto, probabilidad)
    arregloPuntosX.append(punto[0])
    arregloPuntosY.append(punto[1])
    
plt.scatter(np.asarray(arregloPuntosX), np.asarray(arregloPuntosY), s=2, c='green')
plt.show()
```

### Ejercicio 3 - Análisis de pseudorandoms

#### Generador Pseudorandoms

```python
### Se crea la funcion generadora de randoms
def generator(a, m):
    def inner_generator(xo, n):
        items = list(range(n))
        items[0] = xo
        for i in range(1, n):
            items[i] = (a * items[i - 1]) % m
        return [i / float(m) for i in items]
    return inner_generator
```

### Inciso 1

### Generador 1 - Parametros a = 5^5 y m = 2^35 - 1

```python
### Se hace el primer generador con los parametros a = 5^5 y m = 2^35 - 1
generador1 = generator(5**5,(2**35) - 1)
```

#### 100 repeticiones de randoms

```python
### Ponemos un xo inicial y la cantidad de repeticiones para generar randoms
xo11 = 3
n11 = 100
items = generador1(xo11, n11)

plt.hist(items, bins=10)
plt.show()
```
